```python
import pandas as pd
import numpy as np

# Load the CSV file
df = pd.read_csv(r"c:\Users\hanss\git\DSCwashumed\backend\data\Review_SY-08002944_4_3_2025 10_31_21.csv")

# Define the desired column order
desired_order = [
    "Sample ID", "Patient ID", "Patient", "WBC (10^3/uL)", "Neu # (10^3/uL)", "Lym # (10^3/uL)",
    "Mon # (10^3/uL)", "Eos # (10^3/uL)", "Bas # (10^3/uL)", "Neu % (%)", "Lym % (%)", "Mon % (%)",
    "Eos % (%)", "Bas % (%)", "RBC (10^6/uL)", "HGB (g/dL)", "HCT (%)", "MCV (fL)", "MCH (pg)",
    "MCHC (g/dL)", "RDW-CV (%)", "PLT (10^3/uL)", "MPV (fL)", "Species", "Sample State",
    "Owner Last Name", "Mode", "Date", "Time", "Gender", "Age", "Ref. Group", "Draw Date",
    "Draw Time", "Delivery Date", "Delivery Time", "Veterinarian", "Operator", "Comments",
    "WBC Message", "RBC Message", "PLT Message"
]

# Reorder columns (ignore missing columns)
df = df[[col for col in desired_order if col in df.columns]]

# Save the cleaned data
df.to_csv(r"c:\Users\hanss\git\DSCwashumed\backend\data\Review_SY-08002944_4_3_2025 10_31_21_cleaned.csv", index=False)

# Show the cleaned table
display(df)

# Table of missing data per column
missing_data = pd.DataFrame({
    'Missing Count': df.isnull().sum(),
    'Missing %': df.isnull().mean() * 100
})
display(missing_data)
```
